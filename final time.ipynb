```python
from tkinter import *
import datetime
import time
from plyer import notification

def Notification():
    alarmHour=int(tem.get())
    alarmMinute =int(temp.get())
    ampm=str(tempp.get())
    if (ampm=="pm"):
        alarmHour=alarmHour+12
        
    while(1==1) :
        if(alarmHour==datetime.datetime.now().hour and 
           alarmMinute==datetime.datetime.now().minute):
            i=0
            while i<1:
                notification.notify(
                title = "**LETS GO FOR DATE!!",
                message ="Hurry Up This is a Friendly Reminder that you will only get a limited time to create an impression,don't let ur date wait ,make them happy by getting them a surprise gift for this VALENTINE",
                app_icon ="F:\choc.ico",
                timeout= 5
                )
                i+=1
        
            break
    print("exit")  




def clear_handler():
    tem.set(" ")
    temp.set(" ")
    tempp.set(" ")

window=Tk()
tem=StringVar()
temp=StringVar()
tempp=StringVar()


label2=Label(window,text="Valentine Reminder", width= 20, font= ("Pacifico","20"), fg=
'red')
label2.place(x=150,y=30)

lbl=Label(window, text="Hour:", width= 13, font= ("Helvetica","10"), fg=
'blue')
lbl.place(x=140, y=100)
ent=Entry(window, width=20,textvariable=tem, fg='blue')
ent.place(x=270, y=100)

lbl2=Label(window, text="Minute:", width= 13, font= ("Helvetica","10"), fg=
'blue')
lbl2.place(x=140, y=135)
ent2=Entry(window, width=20,textvariable=temp, fg='blue')
ent2.place(x=270, y=135)

lbl3=Label(window, text="Am/Pm:", width= 13, font= ("Helvetica","10"), fg=
'blue')
lbl3.place(x=140, y=170)
ent3=Entry(window, width=20,textvariable=tempp, fg='blue')
ent3.place(x=270, y=170)



btn=Button(window,text="SET", width=10, height=2, font=("Helvetica","10"),
bg='LightCoral',fg='blue',command=Notification)
btn.place(x=250, y=250)



btn_clear=Button(window, text="Clear ALL", width=10, height=2,
font=("Helvetica","10"), bg='pink',fg='blue',command=clear_handler)
btn_clear.place(x=250, y=320)

window.title('Set Your Time!!!')
window.configure(background="crimson")
window.geometry("300x200+10+10")

window.maxsize(width=900,height=700)
window.minsize(width=600,height=600)

window.mainloop()
```
